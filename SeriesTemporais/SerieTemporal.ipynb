```python
import numpy as np
import matplotlib.pyplot as plt

m = 1

serie = np.genfromtxt("Datasets/ptbdb_normal.csv", delimiter=',',skip_header=0,max_rows=m)

for i in range(m):
    if m == 1:
        plt.plot(serie)
    else:
        plt.plot(serie[i])
plt.show()
```

```python
from pyts.image import GramianAngularField
from pyts.image import MarkovTransitionField
from pyts.image import RecurrencePlot

if m == 1:
    serie = np.reshape(serie,(1,-1))
SerieGaf = GramianAngularField().fit_transform(serie)
SerieMtf = MarkovTransitionField().fit_transform(serie)
SerieRecur = RecurrencePlot().fit_transform(serie)

i = 0

print(str(SerieGaf[0]))

plt.subplot(1,3,1)
plt.imshow(SerieGaf[i], cmap='rainbow', origin='lower', vmin=0., vmax=1.)
plt.subplot(1,3,2)
plt.imshow(SerieMtf[i], cmap='rainbow', origin='lower', vmin=-1., vmax=1.)
plt.subplot(1,3,3)
plt.imshow(SerieRecur[i], cmap='rainbow', origin='lower', vmin=-1., vmax=1.)
plt.show()
```
